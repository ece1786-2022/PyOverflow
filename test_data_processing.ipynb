```python
# import csv files from data/test_data
import pandas as pd

data = pd.read_csv('data/Test/regex.csv')
data.head()
```

```python
# remove all the columns except for id, body, title and tags
data = data[['Id', 'Body', 'Title', 'Tags']]
data.head()
```

```python
# remove everything in between <code> and </code> tags
# remove all the html tags
import re
def remove_code_tags(text):
	    return re.sub(r'<code>.*?</code>', '', text, flags=re.DOTALL)

data['Body'] = data['Body'].apply(remove_code_tags)
data.head()
```

```python
data['Body'] = data['Body'].str.replace('<.*?>', '')

# remove all /n, /t, /r, /u characters
data['Body'] = data['Body'].str.replace(r'[\n\t\r]', ' ')
```

```python
data.head()
```

```python
# save the processed data to a csv file
data.to_csv('data/Test/regex.csv', index=False)
```
